# Find Percentile Rank

```python
def PercentileRank(scores, your_score):
    count = 0
    for score in scores:
        if score <= your_score:
            count += 1

    percentile_rank = 100.0 * count / len(scores)
    return percentile_rank
```

computing the total percentile rank of 2 variables:
1.URBAN total population 2011
2.RURAL population 2011

PercentileRank takes a value and computes its percentile rank in a set of values

# Find Percentile

Percentile takes a percentile rank and computes the corresponding percentile. 

```python
def Percentile2(scores, percentile_rank):
    scores.sort()
    # // Divides and returns the integer value of the quotient. It dumps the digits after the decimal.
    index = percentile_rank * (len(scores)-1) // 100
    return scores[index]
```

# Calculating CDF

The CDF is the function that maps from a value to its percentile rank.

```python
def EvalCdf(sample, x):
    count = 0.0
    for value in sample:
        if value <= x:
            count += 1

    prob = count / len(sample)
    return prob
```

```python
t = [68,60,60,69,60,66,87,72,71,82,63,61,69,93,64,77,76,70,84,75,70,71,61,79,74,85,66,76,72,85,81,64,82,82,92,80]
```

```python
EvalCdf(t, 68), EvalCdf(t, 60), EvalCdf(t, 69), EvalCdf(t, 87), EvalCdf(t, 72), EvalCdf(t, 71)
```

# Representing CDF

thinkstats2 provides a class named Cdf that represents CDFs. The fundamental methods Cdf provides are:

    Prob(x):
    Given a value x, computes the probability p = CDF(x). The bracket operator is equivalent to Prob.

    Value(p):
    Given a probability p, computes the corresponding value, x; that is, the inverse CDF of p.


```python

cdf = thinkstats2.Cdf(t, label='prglngth')
thinkplot.Cdf(cdf)

thinkplot.Config(xlabel='kglj', ylabel='CDF', loc='upper left')
```

For example, 86.1% of pregnancy lengths are less than or equal to 84.

```python
cdf.Prob(84)
```

Value evaluates the inverse CDF; given a fraction, it computes the corresponding value. For example, the median is the value that corresponds to 0.5.


```python
cdf.Value(0.5)
```

The CDF also provides a visual representation of the shape of the distribution. Common values appear as steep or vertical sections of the CDF; in this example, the mode and median at 72 Literacy rate value is apparent.
